#### Chart - 5

```python
# Chart - 5 visualization code

window_length = 14
delta = df['Close'].diff()
up, down = delta.clip(lower=0), -delta.clip(upper=0)
roll_up = up.rolling(window=window_length).mean()
roll_down = down.rolling(window=window_length).mean()
RS = roll_up / roll_down
RSI = 100 - (100 / (1 + RS))

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], RSI, color='purple')
plt.title('Relative Strength Index (RSI)')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.grid()
plt.show()
```

##### 1. Why did you pick the specific chart?

RSI measures momentum and identifies overbought/oversold conditions.

##### 2. What is/are the insight(s) found from the chart?

RSI > 70 suggests overbought, while RSI < 30 indicates oversold conditions.

#### Chart - 10

```python
# Chart - 10 visualization code

rolling_mean = df['close'].rolling(window=20).mean()
rolling_std = df['close'].rolling(window=20).std()
upper_band = rolling_mean + (rolling_std * 2)
lower_band = rolling_mean - (rolling_std * 2)
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['close'], label='Close Price', color='blue')
plt.plot(df['date'], rolling_mean, label='20-Day SMA', color='orange')
plt.fill_between(df['date'], lower_band, upper_band, color='gray', alpha=0.3)
plt.title('Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.show()
```

##### 1. Why did you pick the specific chart?

Bollinger Bands help detect overbought and oversold conditions.

##### 2. What is/are the insight(s) found from the chart?

When prices touch the upper band, the stock may be overbought, while hitting the lower band suggests an oversold state.
